## 5. Solve following question by using qcut(10)
### 1. Enrollment Quantile Categories: How are trials distributed across four quantiles based on enrollment numbers?

```python
df['Enrollment Quantile'] = pd.qcut(df['Enrollment'], 4, labels=["Q1", "Q2", "Q3", "Q4"])

enrollment_quantile_distribution = df['Enrollment Quantile'].value_counts().sort_index()


enrollment_quantile_distribution
```

### 2. Start Date Year Quantiles: What are the time periods represented by four quantiles of trial start years?


```python
df['Start Year'] = df['Start Date'].dt.year

df['Start Year']
```

```python
quartile_labels = ['Q1', 'Q2', 'Q3', 'Q4']
df["start_quartiles_custom"] = pd.qcut(df["Start Date"], q=4, labels=quartile_labels)
df[["Start Year", "start_quartiles_custom"]]
```

```python
df["start_quartiles_custom"].value_counts()
```

### 3. Completion Time Frame Categories: How do trial durations distribute across four quantiles from start to completion date?

```python
df[['Start Date', 'Completion Date']]
```

```python
df['Completion Date'] = pd.to_datetime(df['Completion Date'], errors='coerce')
df['Completion Date']
```

```python
df['Trial Duration'] = (df['Completion Date'] - df['Start Date']).dt.days
df['Trial Duration']
```

```python
df['Duration Quantile'] = pd.qcut(df['Trial Duration'], 4, labels=["Q1", "Q2", "Q3", "Q4"])
duration_quantile_distribution = df['Duration Quantile'].value_counts().sort_index()
duration_quantile_distribution
```

### 4 Outcome Measures Count Quantiles: How are trials categorized into four quantiles based on the number of outcome measures?

```python

import numpy as np
```

```python
df['Outcome Measures Count'] = df['Outcome Measures'].str.count(";") + 1
```

```python
df[["Outcome Measures Count"]]
```

```python
bins = [0, 5, 10, 15, np.inf]
labels = ['1-5', '6-10', '11-15', '16+']

df['Outcome Measures Category'] = pd.cut(df['Outcome Measures Count'], bins=bins, labels=labels)


outcome_measures_category_distribution = df['Outcome Measures Category'].value_counts().sort_index()

print(outcome_measures_category_distribution)
```

#### 6. Completion Time Frame Grouping: Apply the cut function to categorize the completion time frames of trials into distinct intervals. What are the characteristics of trials within each time frame group?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce', exact=False)
df['Completion Date'] = pd.to_datetime(df['Completion Date'], errors='coerce', exact=False)
```

```python
# Calculate the 'Completion Time Frame' as the difference in days between 'Completion Date' and 'Start Date'
df['Completion Time Frame'] = (df['Completion Date'] - df['Start Date']).dt.days

df['Completion Time Frame']
```

```python

bins = [0, 30, 60, 90, 120, 150, 180, 365, 730, float('inf')]
labels = ['0-30 days', '31-60 days', '61-90 days', '91-120 days', '121-150 days', '151-180 days', '181-365 days', '366-730 days', '>730 days']

# Categorize completion time frames into intervals
df['Time Frame Category'] = pd.cut(df['Completion Time Frame'], bins=bins, labels=labels, include_lowest=True)

# Examine the characteristics of trials within each time frame group
grouped = df.groupby('Time Frame Category').describe()

# Print the grouped data to see characteristics of trials within each time frame group
print(grouped)
```

#### 7. Enrollment Distribution Analysis: Use Matplotlib&#39;s hist() function to create a histogram visualizing the distribution of the &#39;Enrollment&#39; column in the dataset.

```python
df['Enrollment'] = pd.to_numeric(df['Enrollment'], errors='coerce')

df_cleaned_for_enrollment = df.dropna(subset=['Enrollment'])

plt.figure(figsize=(10, 6))
plt.hist(df_cleaned_for_enrollment['Enrollment'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Enrollment')
plt.xlabel('Enrollment')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
